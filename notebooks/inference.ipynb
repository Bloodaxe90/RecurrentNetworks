```python
import matplotlib.pyplot as plt

def plot_results(*args, x_plot: str = "Time", title: str = "", labels: tuple = ()):
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))
    colours = []

    for idx, result in enumerate(args):
        label = f"r{idx}" if len(labels) - 1 < idx else f"{labels[idx]}"
        x = result.index if x_plot == "index" else result[x_plot]

        colour = plt.cm.hsv(idx / len(args))
        colours.append(colour)

        ax[0].plot(x, result["Loss"], label=label, color=colour)
        ax[1].plot(x, result["Accuracy"], label=label, color=colour)

        ax[0].axhline(result["Loss"].mean(), label=f"{label} Mean Loss", linestyle="--", color=colour)
        ax[1].axhline(result["Accuracy"].mean(), label=f"{label} Mean Accuracy", linestyle="--", color=colour)

    ax[0].set_xlabel(x_plot)
    ax[0].set_ylabel("Loss")
    ax[0].legend()

    ax[1].set_xlabel(x_plot)
    ax[1].set_ylabel("Accuracy")
    ax[1].legend()
    if title != "":
        fig.suptitle(title, fontsize=16, fontweight="bold")
    plt.show()
```

```python
date_dir = "2025-04-10"

b_r_results = load_results(date_dir= date_dir,
             exp_name= "baseline_Fashion_MNIST_RecurrentLayer_NL2_LR0.0001_E2_B64_ID28_HD64_OD10_SD32_STrue_DP0.2.csv")
b_lstm_results = load_results(date_dir= date_dir,
             exp_name= "baseline_Fashion_MNIST_LSTMLayer_NL2_LR0.0001_E2_B64_ID28_HD64_OD10_SD32_STrue_DP0.2.csv")
b_gru_results = load_results(date_dir= date_dir,
             exp_name= "baseline_Fashion_MNIST_GRULayer_NL2_LR0.0001_E2_B64_ID28_HD64_OD10_SD32_STrue_DP0.2.csv")

r_results = load_results(date_dir= date_dir,
             exp_name= "frfr_Fashion_MNIST_RecurrentLayer_NL2_LR0.0001_E2_B64_ID14_HD64_OD10_SD32_STrue_DP0.2.csv")
lstm_results = load_results(date_dir= date_dir,
             exp_name= "frfr_Fashion_MNIST_LSTMLayer_NL2_LR0.0001_E2_B64_ID14_HD64_OD10_SD32_STrue_DP0.2.csv")
gru_results = load_results(date_dir= date_dir,
             exp_name= "frfr_Fashion_MNIST_GRULayer_NL2_LR0.0001_E2_B64_ID14_HD64_OD10_SD32_STrue_DP0.2.csv")

plot_results(b_r_results, b_lstm_results, b_gru_results, title="Baseline training results for input of shape (batch, 28, 28)", labels= ("Recurrent", "LSTM", "GRU"))
plot_results(r_results, lstm_results, gru_results, title= "Final training results for input of shape (batch, 56, 14)", labels= ("Recurrent", "LSTM", "GRU"))
```
